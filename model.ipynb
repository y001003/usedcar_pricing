# 2. EDA/ Data Preprocessing

```python
avante['car'] = 'avante'
genesis['car'] = 'genesis'
grandure['car'] = 'grandure'
sonata['car'] = 'sonata'
santafe['car'] = 'santafe'
avante
```

```python
avante.info(),genesis.info(),grandure.info(),sonata.info(),santafe.info()
```

```python
# 전처리 함수
def cleaning(df) :
    df['odometer'] = df['odometer'].str.replace(',','').str.replace('km','')
    df['price'] = df['price'].str.replace(',','')
    df['price'] = df['price'].str.replace(r'\D', '') # price에 string값들 지우기
    df['odometer'] = df['odometer'].str.replace(r'\D', '') # odometer에 string값들 지우기
    df.drop(df[df['price'] == ''].index,axis=0,inplace=True) # 지운 것 포함 row 삭제
    df.drop(df[df['odometer'] == ''].index,axis=0,inplace=True) # 지운 것 포함 row 삭제
    df.drop(df[df['price'].isnull()].index,axis=0,inplace=True) # 결측치 포함 row 삭제
    df['price'] = df['price'].astype('int') # int로 변환
    df['odometer'] = df['odometer'].astype('int') # int로 변환
    return df
```

```python
avante_df = avante.copy()
genesis_df = genesis.copy()
grandure_df = grandure.copy()
sonata_df = sonata.copy()
santafe_df = santafe.copy()
```

```python
avante_df = cleaning(avante_df)
genesis_df = cleaning(genesis_df)
grandure_df = cleaning(grandure_df)
sonata_df = cleaning(sonata_df)
santafe_df = cleaning(santafe_df)
```
